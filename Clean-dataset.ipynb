```python
import hashlib
def md5(file_path):
    with open(file_path, 'rb') as f:
        hash_md5 = hashlib.md5()
        for chunk in iter(lambda: f.read(4096), b""):
            hash_md5.update(chunk)
    return hash_md5.hexdigest()
```

```python
for key in hash_dict2.keys():
    file_path = hash_dict1[key]
    src_path = os.path.join('data/plantains', iu.get_file_or_foldername(file_path).replace('jpg', 'png'))
    dest_path = os.path.join(
        'data/labeled-plantain', iu.get_file_or_foldername(hash_dict2[key]).replace('jpg', 'png'))
    shutil.copy2(src_path, dest_path)
    print('copied', src_path, 'to', dest_path)
```
